```python
# Dependencies
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Read CSV
got_data = pd.read_csv("Resources/got.csv")
got_data.head()
```

```python
# Get attacker king data
attacker_data = got_data["attacker_king"].value_counts()
attacker_data
```

```python
# Get defender king data
defender_data = got_data["defender_king"].value_counts()
defender_data
```

```python
battle_df = pd.DataFrame({
    "Attacker": attacker_data,          
    "Defender": defender_data 
})

battle_df = battle_df.fillna(0)

battle_df
```

```python
# Create bar chart (of sorted values)
battle_df.sort_values(by="Attacker").plot.barh() #stacked=True
#battle_df.sort_values(by="Attacker").plot.barh(y=["Attacker", "Defender"])
#battle_df.sort_values(by="Attacker").plot(kind="barh")

# Set textual properties
plt.title("The Bloodthirst of Kings")
plt.xlabel("Number of Battles Participated In")
#plt.ylabel("King")

# Resize plot to display labels
plt.tight_layout()

# Show plot
plt.show()
```
